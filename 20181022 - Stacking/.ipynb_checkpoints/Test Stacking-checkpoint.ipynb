```python
from sklearn.datasets import load_digits
import sys
sys.path.insert(0, '../SOM-LVQ')
import sklearn_LVQ
from stacked_generalizer import StackedGeneralizer
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

VERBOSE = True
N_FOLDS = 5

# load data and shuffle observations
data = load_digits()

X = data.data
y = data.target

shuffle_idx = np.random.permutation(y.size)

X = X[shuffle_idx]
y = y[shuffle_idx]

# hold out 20 percent of data for testing accuracy
train_prct = 0.8
n_train = int(round(X.shape[0]*train_prct))

# define base models
base_models = [sklearn_LVQ.LVQ(x = X[:n_train], y = y[:n_train], n_classes=9, p_vectors=[], n_neurons=35),
               sklearn_LVQ.LVQ(x = X[:n_train], y = y[:n_train], n_classes=9, p_vectors=[], n_neurons=30),
               sklearn_LVQ.LVQ(x = X[:n_train], y = y[:n_train], n_classes=9, p_vectors=[], n_neurons=25)]

# define blending model
# blending_model = LogisticRegression()
blending_model = KNeighborsClassifier()

# initialize multi-stage model
sg = StackedGeneralizer(base_models, blending_model, 
	                    n_folds=N_FOLDS, verbose=VERBOSE)

# fit model
sg.fit(X[:n_train],y[:n_train])

# test accuracy
pred = sg.predict(X[n_train:])
pred_classes = [np.argmax(p) for p in pred]

_ = sg.evaluate(y[n_train:], pred_classes)
```
